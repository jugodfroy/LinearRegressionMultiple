#### Import the data, clean it, and prepare the input and output vectors

```python
x, y, df = import_clean_data('./data/Auto.csv', ['horsepower'], ['mpg'])
X, y = prepare_vectors(x, y)
```

#### For each degree, compute the regression  

```python
degrees_list = [1, 2, 5]

model_list = []
for i, degree in enumerate(degrees_list):
    model = Model(X, y, learning_rate=0.0008, iterations=6000, test_size=0.2)   #Create the model with the chosen parameters
    model.transform(degree)             #IMPORTANT STEP : Transforming X vector into polynomial vector  
    model.compute_regression()
    model_list.append(model)
    
```

#### For each model, compute the regression and plot the results and r2 score

```python

for i, model in enumerate(model_list):
    model.compute_r_square()
    model.plot_regression_2D('horsepower', 'mpg', 'regression (degree = {})'.format(degrees_list[i]))

```
